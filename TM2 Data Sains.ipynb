```python
import pandas as pd
import numpy as np
df = pd.read_csv("datatni.csv")

# Kodingan Menghitung statistik lanjutan, .2f buat mastiin hasil dua angka dibelakang koma
std_dev = df["tinggi"].std()
variance = df["tinggi"].var()

print("\n=== Statistik Lanjutan ===")
print(f"Standar Deviasi : {std_dev:.2f}")
print(f"Variansi        : {variance:.2f}")

# Kodingan Cari calon dengan tinggi badan tertinggi
tertinggi = df.loc[df["tinggi"].idxmax()]
print("\n=== Calon dengan Tinggi Badan Tertinggi ===")
print(f"Nama    : {tertinggi['name']}")
print(f"Alamat  : {tertinggi['alamat']}")
print(f"Tinggi  : {tertinggi['tinggi']} cm")
```
