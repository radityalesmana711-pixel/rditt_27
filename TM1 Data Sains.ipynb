```python
import pandas as pd

data = {
    "Nama": ["Radit", "Reva", "Depi", "Ridwan", "Nabilla", "Aufa", "Erina", "Fikri", "Dastanta", "Ressy"],
    "Matematika": [80, 75, 90, 65, 85, 70, 95, 60, 88, 77],
    "Fisika": [78, 82, 85, 70, 80, 75, 92, 68, 83, 79],
    "Kimia": [88, 74, 91, 69, 84, 72, 96, 63, 85, 81],
    "Biologi": [85, 79, 88, 73, 82, 76, 93, 70, 86, 80]
}

df = pd.DataFrame(data)

# Menghitung rata-rata nilai setiap siswa
df["Rata-rata"] = df[["Matematika", "Fisika", "Kimia", "Biologi"]].mean(axis=1)

# Mengurutkan berdasarkan nilai rata-rata (descending)
df_sorted = df.sort_values(by="Rata-rata", ascending=False)

# Menambahkan kolom peringkat
df_sorted["Peringkat"] = range(1, len(df_sorted) + 1)

# Menampilkan hasil akhir
print("Data Nilai Siswa dengan Rata-rata dan Peringkat:\n")
print(df_sorted)
```
